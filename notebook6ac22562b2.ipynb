## Univariate Analysis 

### For Numerical Features 

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select numerical columns
numerical_columns = df.select_dtypes(include=['number']).columns

# Create subplots: Determine grid size
num_cols = len(numerical_columns)
cols = 3  # Number of plots per row
rows = (num_cols + cols - 1) // cols  # Calculate the number of rows needed

# Create a figure with subplots
fig, axes = plt.subplots(rows, cols, figsize=(15, rows * 5), constrained_layout=True)  # Use `constrained_layout` for spacing

# Flatten axes for easy iteration (if single row/column, axes might not be nested)
axes = axes.flatten()

# Plot each numerical column
for i, col in enumerate(numerical_columns):
    sns.histplot(df[col], kde=True, bins=30, color='blue', ax=axes[i])
    axes[i].set_title(f"Distribution of {col}", fontsize=14)
    axes[i].set_xlabel(col, fontsize=12)
    axes[i].set_ylabel("Frequency", fontsize=12)

# Remove unused subplots if any
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Show the plot
plt.show()
```

```python
# Select categorical columns
categorical_columns = df.select_dtypes(include=['object']).columns

# Determine grid size
num_cols = len(categorical_columns)
cols = 3  # Number of plots per row (you can adjust this)
rows = (num_cols + cols - 1) // cols  # Calculate the number of rows needed

# Create a figure with subplots
fig, axes = plt.subplots(rows, cols, figsize=(15, rows * 6), constrained_layout=True)

# Flatten axes for easy iteration
axes = axes.flatten()

# Plot each categorical column
for i, col in enumerate(categorical_columns):
    sns.countplot(data=df, y=col, order=df[col].value_counts().index, palette="viridis", ax=axes[i])
    axes[i].set_title(f"Frequency of {col}", fontsize=14)
    axes[i].set_xlabel("Count", fontsize=12)
    axes[i].set_ylabel(col, fontsize=12)

# Remove unused subplots if any
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Show the plot
plt.show()
```

## Using Neural Network

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import r2_score, f1_score, classification_report
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
# Build a simple neural network
model = Sequential()
model.add(Dense(256, activation='relu', input_shape=(x_train.shape[1],)))
model.add(Dense(128, activation='relu'))
model.add(Dense(1, activation="sigmoid"))
```

```python
# Compile the model with F1 metric
model.compile(optimizer='adam', 
              loss= 'binary_crossentropy',
              metrics=['accuracy'])

# Early stopping to avoid overfitting
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

# Train the model
history = model.fit(
    x_train_scaled, y_train,
    validation_data=(x_test_scaled, y_test),
    epochs=100,
    batch_size=32,
    callbacks=[early_stop],
    verbose=1
)
```

```python
y_pred_probs = model.predict(x_test_scaled)
y_pred = (y_pred_probs > 0.5).astype(int)
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
print(f"F1 Score: {f1_score(y_test, y_pred, average='weighted')}")

# Plot the training history (optional)
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.title('Training History')
plt.show()
```
